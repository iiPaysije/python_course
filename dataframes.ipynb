```python
import pandas

df = pandas.read_json('supermarkets.json')
df
```

```python
df["Address"] = df["Address"] + "," + df["City"] + "," + df["State"]
df
```

```python
df["Location"] = df["Address"].apply(nom.geocode)

# Using inline lamda function with simple if else to prevent errors
df["Latitude"] = df["Location"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"] = df["Location"].apply(lambda x: x.longitude if x != None else None)
df
```
